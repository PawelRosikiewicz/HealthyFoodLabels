# ABSTRACT
Nutri-Score is an intuitive system for showing nutritiotional value of food products. It is displayed on food-boxes as so called traffic-lights, with green A letter given to the most heathy food products, and red E letter given to the products commonly known as “junk-food”.  It has been shown up to 51% of consumers use food labels for making their product choices, and the Nutri-Score was adopted or recommended by several Europeans countries and companies, as simple yet often black box guidline for the consumers.  However, the five-level Nutri-score system is based on nutritional grades (-15 to 40) that may be differently calculated in different countries, and it may occcasionally affect Nutri-Score of the same products saled in different countries. 
     
From the above reasons, I compared Nutri-Scores calculated with algorithms used in UK and France for over 60.000 food products that were available in Open Food Facts dataset https://world.openfoodfacts.org. UK and France Nutri-Scores of the same products were compared in 14 or 37 different groups of food products to see whether there are signifficant and consistent changes in information provided to consumers in different countries, and whther these changes may reflect products producced in these countires. 
       
In my report I will show, that relatively small differences in calculating nutritional grades, in UK and France, may strongly affects Nutri-Score labels of very specific product groups such as cheese in France and sweetened beverages in UK, without affecting the labels on food products from many other groups. Moreover, the changes in nutritional grade are often sufficient to present some of these products as more, heathy in earther France or UK, and thus may potentially affect consumer choices.

### Find correlation between variables and nuctrien dcores in products with differen pnns groups

```python
# select the data,
main_group = "nutrition-score-fr_100g" # I compared UK, and FR nutriu scores in Task 2
cols_with_num_data = taskF_helpers.find_columns_with_selected_dtype(df=cleaned_data, dtype="numerical") 
df_subset = cleaned_data.loc[:,cols_with_num_data]

#. remove columns with less then 50% of data
df_subset = taskA_helpers.remove_nan_and_duplicates(df=df_subset, 
                                               col_threshold=50,  # removing columns with >50% of missing data
                                               row_threshold="none", 
                                               drop_duplicates=False, 
                                               verbose=False) # small values just to show that my function works,

# find groups, that you wish to use to divide the data 
group_positions = cleaned_data["pnns_groups_1"]
group_positions = group_positions.str.lower()
group_positions[group_positions.isna()] ="no group"
group_positions = group_positions.str.replace("-"," ")

# get unique values
group_names = list(group_positions.unique())

# find all correlations wiht selected viariable, 
corr_data_in_each_group_dct = dict()
for i, group_name in enumerate(group_names):

    # select the group,
    group_filter    =  list(group_positions==group_name) 
    df_group_subset = df_subset.loc[group_filter,:]    
    
    # prepare the data
    dct_results  = taskF_helpers.clean_data_and_find_correlations(
        df=df_group_subset, 
        colName_main_variable="nutrition-score-fr_100g", 
        verbose=False
        )
    
    # plot data, with correlation values, and display the table with all the results for 10 pairs of variabkles with the highest asb(r) values,
    corr_results = taskF_helpers.make_plots_and_table_with_corr( 
        data_dct=dct_results,
        return_rable=True,
                                                   display_table=False,
                                                   display_plots=False); # iam reusing ths fucntion, just to calculate correlations,
                                                                         # it will not display any plots or tables (False)
    # add to dct
    corr_data_in_each_group_dct[group_name] = corr_results;
    

# heatmap
taskF_helpers.corr_heatmap(corr_res_dct=corr_data_in_each_group_dct, method="rvalues")
# in case you woudl like to see correaltion calcluated with two rank methods, please rung that code:
#.     corr_hist(corr_res_dct=corr_data_in_each_group_dct, method="rho")
#.     corr_hist(corr_res_dct=corr_data_in_each_group_dct, method="tau")
```
